```python
labels = ['강백호', '서태웅', '정대만'] #이름
values = [190, 187, 184] #키

plt.barh(labels, values)
plt.xlim(175, 195)
```

```python
bar = plt.bar(labels, values)
bar[0].set_hatch('/') #///
bar[1].set_hatch('x') # XXXX
bar[2].set_hatch('..') # yyy
```

```python
bar = plt.bar(labels, values)

for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height(), values[idx])
```

```python
bar = plt.bar(labels, values)
plt.ylim(175, 195)

for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height(), values[idx])
```

```python
bar = plt.bar(labels, values)
plt.ylim(175, 195)

for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height() +0.5, values[idx])
```

```python
bar = plt.bar(labels, values)
plt.ylim(175, 195)

for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height() +0.5, values[idx], ha='center', color='blue')
```
