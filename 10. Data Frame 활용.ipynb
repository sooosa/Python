```python
plt.plot(df['지원번호'], df['영어'])
plt.plot(df['지원번호'], df['수학'])

plt.grid(axis='y', color='purple', alpha=0.2, linestyle =':') #  grid에 색깔이 바뀌는 걸 알수있다.alpha = 1 ~ 0.1 투명도를 알려준다.
```

```python
plt.plot(df['지원번호'], df['영어'])
plt.plot(df['지원번호'], df['수학'])

plt.grid(axis='y', color='purple', alpha=0.5, linestyle ='--') # linestyle 변경해주면 라인 변경 된걸 알수있다.
```

```python
plt.plot(df['지원번호'], df['영어'])
plt.plot(df['지원번호'], df['수학'])

plt.grid(axis='y', color='purple', alpha=0.5, linestyle ='--', linewidth=2) # linewidth 기존에선두께보다 더두꺼워지는걸 알수있다. 
```
